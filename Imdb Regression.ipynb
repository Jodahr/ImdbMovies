```python
# Cardinality of categorical columns
df.select_dtypes(include=[object]).nunique()
```

```python
# value counts of categorical columns
print(df.director_name.value_counts().head(),"\n")
print(df.country.value_counts().head(),"\n")
print(df.language.value_counts().head(),"\n")
print(df.color.value_counts().head(),"\n")
print(df.actor_1_name.value_counts().head())
```

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(corr, mask=mask, linewidths=0.5, cmap="coolwarm")
```
